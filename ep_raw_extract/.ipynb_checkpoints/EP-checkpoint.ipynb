```python
import os
import json
import datetime
from astrotate import server
import shutil
import copy
```

```python
root = '/Volumes/OWC Mercury/jzhao electrophysiology/191108'
```

```python
conn = server.connect_server()
```

```python
def readjson(path):
    # read a json file
    with open(path, 'r') as json_file:
        data = json.load(json_file)
    return(data)
```

```python
date = datetime.datetime.strptime(os.path.basename(root), '%y%m%d')
animalid = 'A'+date.strftime('%Y%m%d')+'01'

gender = 'M' # M F
species = 'rat' # rat mouse
strain = 'SD' # SD    C57BL/6J
weight = 317

animalinfo = {}
animalinfo['sub_bleeding']= 'N'
animalinfo['brain_bloom'] = 'N'
animalinfo['give_oxygen'] = 'Y'
treatment = info['treatment']


```

```python
folders = os.listdir(root)
folders = [os.path.join(root, x) for x in folders]
for f in folders[1:2]:
    print(f)
    subfs = os.listdir(f)
    subfs = [os.path.join(f, x) for x in subfs if x[0]!='.']
    print('========================================================================================================')
    print('========================================================================================================')
    print(subfs)
    #if len(subfs)>1:
    #    break
    #sf = subfs[0]
    for sf in subfs:
        print(os.listdir(sf))
        info = readjson(os.path.join(sf, 'info.json'))
        animalid = info['animal']['id']

        for key in info.keys():
            if key not in ['project', 'animal', 'treatment', 'bf', 'single_neuron', 'ECoG', 'lfp', 'multiunit']:
                break

        gender = info['animal']['gender']
        species = info['animal']['species']
        strain = info['animal']['strain']
        weight = int(info['animal']['weight'])

        animalinfo = {}
        try:
            animalinfo['sub_bleeding']= info['animal']['sub_bleeding']
        except:
            pass

        try:
            animalinfo['brain_bloom'] = info['animal']['brain_bloom']
        except:
            pass

        try:
            animalinfo['give_oxygen'] = info['animal']['give_oxygen']
        except:
            pass

        try:
            treatment = info['treatment']
        except:
            treatment = {}


        date = datetime.datetime.strptime(os.path.basename(f), '%y%M%d')

        cur = conn.cursor()
        cur.execute(
        """
        INSERT INTO ep_info (animalid, animalinfo, date, weight, gender, species, strain, treatment)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
        """, (animalid, json.dumps(animalinfo), date, weight, gender, species, strain, json.dumps(treatment))
        )
        conn.commit()

        print('====')
        print('add animal successful')
        cur = conn.cursor()
        cur.execute(
        """
        SELECT * FROM ep_info
        WHERE animalid = '{}'
        """.format(animalid)
        )
        a = cur.fetchall()
        conn.commit()
        print(a)


        # bf ====================================================
        if 'bf' in info.keys():
            bfinfo = info['bf']
            filename = 'bf'+date.strftime('%Y%M%d')+ '0'+animalid[-1]+'.csv'
            srcpath = os.path.join(sf, bfinfo['file_path'])
            trgpath = os.path.join('/Users/Melody/Documents/expdata/EP/BF', filename)
            shutil.copyfile(srcpath, trgpath)
            cur = conn.cursor()
            cur.execute(
            """
            INSERT INTO ep_data_bf (animalid, filepath, id, treat_point)
            VALUES (%s, %s, %s, %s)
            """, (animalid, filename, filename[:-4], json.dumps(bfinfo['treat_point']))
            )
            conn.commit()
            os.remove(srcpath)
            print('insert bf successful')


        # single neuron part ============================================================
        if 'single_neuron' in info.keys():
            for i in range(len(info['single_neuron'])):
                singleinfo = copy.copy(info['single_neuron'][i])
                id = 'n'+date.strftime('%Y%M%d')+'%.02d'%int(singleinfo['id'])
                ap_delay =  singleinfo['neuron_info']['ap_delay']
                neuron_type = singleinfo['neuron_info']['neuron_type']
                spontaneous = singleinfo['neuron_info']['spontaneous']
                ongobaseline = singleinfo['neuron_info']['neuron_ongoing_activity_baseline']


                try:
                    sinus_rf = singleinfo['neuron_info']['sinus_RF']
                except:
                    sinus_rf = None

                try:
                    cortex_rf = singleinfo['neuron_info']['cortex_RF']
                except:
                    cortex_rf = None

                try:
                    th_baseline = singleinfo['neuron_info']['TH_baseline']
                except:
                    th_baseline = None

                try:    
                    sth_baseline = singleinfo['neuron_info']['STH_baseline']
                except:
                    sth_baseline = None

                try:
                    mech_threshold = singleinfo['neuron_info']['mech_threshold']
                except:
                    mech_threshold = None

                try:
                    mech_response_type = singleinfo['neuron_info']['mech_response_type']
                except:
                    mech_response_type = None

                if 'spon' in singleinfo.keys():
                    spon_data = copy.copy(singleinfo['spon'])
                    spon_data['file_path'] = os.path.join('SPON', id+'.csv')
                    shutil.copyfile(os.path.join(sf, singleinfo['spon']['file_path']), 
                                    os.path.join('/Users/Melody/Documents/expdata/EP/SINGLENEURON/', spon_data['file_path']))
                else:
                    spon_data = {}

                if 'mech' in singleinfo.keys():
                    mech_data = copy.copy(singleinfo['mech'])
                    mech_data['file_path'] = os.path.join('MECH', id+'.csv')
                    shutil.copyfile(os.path.join(sf, singleinfo['mech']['file_path']), 
                                    os.path.join('/Users/Melody/Documents/expdata/EP/SINGLENEURON/', mech_data['file_path']))
                else:
                    mech_data = {}

                cur = conn.cursor()
                cur.execute(
                """
                INSERT INTO ep_data_singleneuron (id, animalid, ap_delay, mech_response_type, spontaneous,
                ongoing_activity_baseline, sinus_rf, cortex_rf, neuron_type, mech_threshold, th_baseline, sth_baseline,
                spon_data, mech_data)
                VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
                """, (id, animalid, ap_delay, mech_response_type, spontaneous, ongobaseline, sinus_rf, cortex_rf,
                      neuron_type, mech_threshold, th_baseline, sth_baseline, json.dumps(spon_data), json.dumps(mech_data))
                )
                conn.commit()
                print('insert single neuron successful')

        # lfp part ===================================================
        if 'lfp' in info.keys():
            lfpinfo = info['lfp']
            filename = 'lfp'+date.strftime('%Y%M%d')+ '0'+animalid[-1]+'.csv'
            srcpath = os.path.join(sf, lfpinfo['file_path'])
            trgpath = os.path.join('/Users/Melody/Documents/expdata/EP/LFP', filename)
            shutil.copyfile(srcpath, trgpath)
            cur = conn.cursor()
            cur.execute(
            """
            INSERT INTO ep_data_lfp (animalid, filepath, id, treat_point, scanrate)
            VALUES (%s, %s, %s, %s, %s)
            """, (animalid, filename, filename[:-4], json.dumps(lfpinfo['treat_point']), lfpinfo['fq'])
            )
            conn.commit()
            os.remove(srcpath)
            print('insert lfp successful')

            # lfp part ===================================================

        if 'ECoG' in info.keys():
            ecoginfo = info['ECoG']
            filename = 'ecog'+date.strftime('%Y%M%d')+ '0'+animalid[-1]+'.csv'
            srcpath = os.path.join(sf, ecoginfo['file_path'])
            trgpath = os.path.join('/Users/Melody/Documents/expdata/EP/ECOG', filename)
            shutil.copyfile(srcpath, trgpath)
            cur = conn.cursor()
            cur.execute(
            """
            INSERT INTO ep_data_ecog (animalid, filepath, id, treat_point, scanrate)
            VALUES (%s, %s, %s, %s, 1000)
            """, (animalid, filename, filename[:-4], json.dumps(ecoginfo['treat_point']))
            )
            conn.commit()
            os.remove(srcpath)
            print('insert ECoG successful')

        if 'multiunit' in info.keys():
            multiinfo = info['multiunit']
            filename = 'multi'+date.strftime('%Y%M%d')+ '0'+animalid[-1]+'.csv'
            srcpath = os.path.join(sf, multiinfo['file_path'])
            trgpath = os.path.join('/Users/Melody/Documents/expdata/EP/MULTIUNIT', filename)
            shutil.copyfile(srcpath, trgpath)
            cur = conn.cursor()
            cur.execute(
            """
            INSERT INTO ep_data_multiunit (animalid, filepath, id, treat_point, scanrate)
            VALUES (%s, %s, %s, %s, 1)
            """, (animalid, filename, filename[:-4], json.dumps(multiinfo['treat_point']))
            )
            conn.commit()
            os.remove(srcpath)
            print('insert multiunit successful')

        print('delete %s' % sf)
        shutil.rmtree(sf)
        if len(os.listdir(os.path.dirname(sf))) == 0:
            print('delete %s' % os.path.dirname(sf))
            shutil.rmtree(os.path.dirname(sf))
```
