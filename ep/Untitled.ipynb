```python
import numpy as np
import os
import scipy.signal as signal
import matplotlib.pyplot as plt

file_path = os.path.join('/Users/Melody/OneDrive/lab_data/data/LFP/raw', 'C180410011.csv')
# file_path = 'example.csv'
data = np.genfromtxt(file_path)

# np.savetxt("example.csv", data[0:10000], delimiter=",")
```

```python
data1 = np.reshape(data, (4, -1))
```

```python
def psd(array, fs = 1000, nperseg = 1000):
    (r,c) = np.shape(array)
    for i in range(r):
        f, psd = signal.welch(array[i,:], fs = fs, nperseg = nperseg)
        plt.plot(psd)
    plt.show()
```

```python
psd(data1)
```

```python
f, psd = signal.welch(data, 1000, nperseg=1000)
fx, t, sxx = signal.spectrogram(data, fs = 1000, nperseg = 1000)
```

```python
sxxx = sxx[0:50, :]
fxxx = fx[0:50]
```

```python
plt.plot(psd)
```

```python
plt.semilogy(f,psd,'k') # I don't think I should use log 
```

```python
plt.pcolormesh(t, fxxx, sxxx, cmap=plt.cm.get_cmap('jet'))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
